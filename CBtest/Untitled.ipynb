```python
import numpy as np
import matplotlib.pylab as plt
from matplotlib import colors
```

```python
fig=plt.figure(figsize=(10,6))
grid = plt.GridSpec(1, 2, wspace=0.0, hspace=0.0)



x=y=np.linspace(-np.pi,np.pi,300)
XX,YY = np.meshgrid(x,y)
ZZ1=XX + YY
levels_conf = np.array([ZZ1.min(),ZZ1.max()])

ax1 = fig.add_subplot(grid[0, 0])
ax1.ticklabel_format(style='sci', axis='y', scilimits=(0,0),useMathText=True)
CS1 = ax1.imshow(ZZ1,interpolation='bilinear',origin='lower',extent=[XX.min(),XX.max(),YY.min(),YY.max()],vmin=levels_conf.min(), vmax=levels_conf.max(),cmap='jet')
ax1.label_outer()



ax2 = fig.add_subplot(grid[0, 1],sharey=ax1,sharex=ax1)
ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0),useMathText=True)
ZZ2=XX + 2.0*YY
levels_conf = np.array([ZZ1.min(),ZZ1.max()])
CS2 = ax2.imshow(ZZ2,interpolation='bilinear',origin='lower',extent=[XX.min(),XX.max(),YY.min(),YY.max()],vmin=levels_conf.min(), vmax=levels_conf.max(),cmap='jet')
ax2.label_outer()


axes = np.array(fig.get_axes())
print axes
print type(axes)

#ax0 = fig.add_axes(grid[0, :])
#from mpl_toolkits.axes_grid1 import make_axes_locatable
#ax2_divider = make_axes_locatable(axes)
#cax2 = ax2_divider.append_axes("top", size="7%", pad="2%")
#cb2 = plt.colorbar(CS1, cax=cax2, orientation="horizontal",pad = 0.2,fraction = 0.002)
#cax2.xaxis.set_ticks_position("top")


#cbar=fig.colorbar(CS1,ax=axes.ravel().tolist(),shrink=0.95,format="%1.2e",orientation = 'horizontal',fraction=0.07,anchor=(0.0,1.0))
#plt.colorbar(CS1,ax = ax0, use_gridspec=True)
    
cb_ax = fig.add_axes([0.0, 0.9, 1.0,0.05])
#cbar = grid.cbar_axes[0].colorbar(im)
cbar = fig.colorbar(CS1, cax=cb_ax,orientation = 'horizontal',shrink=0.3)
cbar.ax.xaxis.set_ticks_position('top')
cbar.ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.subplots_adjust(wspace=0.0,hspace = 0.0)
plt.show()
```
