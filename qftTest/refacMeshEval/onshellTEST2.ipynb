```python
from __future__ import print_function
import numpy as np
import sftrident.coordinate as coord
import sftrident.util as util
from IPython.display import display, clear_output
```

```python
def transfrom(x,a,b):
    return a + (b-a)*x
    
    
def transfromAll(co):
    ss= np.array([3.353]*len(co))
    p1m = transfrom(co[:,0],1e-4,1.5)
    p1x = transfrom(co[:,1],-10,10)
    p1y = transfrom(co[:,2],-10,10)
    p2m = transfrom(co[:,3],1e-4,1.5)
    p2x = transfrom(co[:,4],-10,10)
    p2y = transfrom(co[:,5],-10,10)
    return np.column_stack([ss,p1m,p1x,p1y,p2m,p2x,p2y])
    

def build_rand_arr(n):
    return np.random.rand(n,6)
```

```python
def physArea_cond_single(pm,px,py,m,atol=1e-5,rtol=1e-4):
    """
    condition for momenta in light cone coordinates
    """
    pp = (px**2 + py**2 + m**2)/(4.0*pm)
    rawMsq = (pp+pm)**2 - px**2 - py**2 - (pp-pm)**2
    
    posPM = (pm>atol)
    pmPos,pxPos,pyPos = [pm[posPM],px[posPM],py[posPM]]
    ppPos = (pxPos**2 + pyPos**2 + m**2)/(4.0*pmPos)
    massSQ = (ppPos+pmPos)**2 - pxPos**2 - pyPos**2 - (ppPos-pmPos)**2
    onshell = (np.abs(massSQ - m**2) < rtol * m**2)
    posPM[np.where(posPM)] = onshell
    return posPM,massSQ,pm,px,py
```

```python
task = 0
eps = 1e-3
breakPoint=0
while True:
    if breakPoint>=4:
        break
    else:
        task+=1
        clear_output(wait=True)
        print("task",task)
        
        
    lccArr = transfromAll(build_rand_arr(100000))

    #print(lccArr.shape)

    testMom = coord.momenta(lccArr,mode='batch')
    
    #test P1
    
    m1sq = testMom.P1*testMom.P1
    lccGrid1 = testMom.grid.lccGrid3
    
    onshell1 = (np.abs(m1sq - 1.0)<eps)
    physArea1,physMass1,p1m,p1x,p1y = physArea_cond_single(*lccGrid1,m=1.0,atol=0.0,rtol=eps)
    
    
    
    m2sq = testMom.P2*testMom.P2
    lccGrid2 = testMom.grid.lccGrid2
    
    onshell2 = (np.abs(m2sq - 1.0)<eps)
    physArea2,physMass2,p2m,p2x,p2y = physArea_cond_single(*lccGrid2,m=1.0,atol=0.0,rtol=eps)
    
    
    
    m3sq = testMom.P3*testMom.P3
    lccGrid3 = testMom.grid.lccGrid1
    
    onshell3 = (np.abs(m3sq - 1.0)<eps)
    physArea3,physMass3,p3m,p3x,p3y = physArea_cond_single(*lccGrid3,m=1.0,atol=0.0,rtol=eps)
    
    for i,el in enumerate(onshell1):
        if el!=physArea1[i]:
            print("==== ERROR1")
            print(el,m1sq[i])
            print(physArea1[i],physMass1[i])
            print("\t",p1m[i],p1x[i],p1y[i])
            print(lccGrid1[0][i],lccGrid1[1][i],lccGrid1[2][i])
            print(lccGrid2[0][i],lccGrid2[1][i],lccGrid2[2][i])
            print(lccGrid3[0][i],lccGrid3[1][i],lccGrid3[2][i])
            breakPoint+=1
    
    for i,el in enumerate(onshell2):
        if el!=physArea2[i]:
            print("==== ERROR2")
            print(el,m2sq[i])
            print(physArea2[i],physMass2[i])
            print("\t",p2m[i],p2x[i],p2y[i])
            print(lccGrid1[0][i],lccGrid1[1][i],lccGrid1[2][i])
            print(lccGrid2[0][i],lccGrid2[1][i],lccGrid2[2][i])
            print(lccGrid3[0][i],lccGrid3[1][i],lccGrid3[2][i])
            breakPoint+=1
    
    for i,el in enumerate(onshell3):
        if el!=physArea3[i]:
            print("==== ERROR3")
            print(el,m3sq[i])
            print(physArea3[i],physMass3[i])
            print("\t",p3m[i],p3x[i],p3y[i])
            print(lccGrid1[0][i],lccGrid1[1][i],lccGrid1[2][i])
            print(lccGrid2[0][i],lccGrid2[1][i],lccGrid2[2][i])
            print(lccGrid3[0][i],lccGrid3[1][i],lccGrid3[2][i])
            breakPoint+=1
    
    
    
    
    #if task == 1:
    #    break
    
    #task+=1
    #clear_output(wait=True)
    #print("task",task)
    
```
