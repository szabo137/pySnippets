```python
import numpy as np
from scipy import interpolate
import matplotlib.pylab as plt
plt.style.use("seaborn")
```

```python
x=np.linspace(-10*np.pi,10*np.pi,100)
y=np.linspace(-10*np.pi,10*np.pi,100)

XX,YY = np.meshgrid(x,y)
ZZ=np.cos(XX) + np.sin(YY/3)
print XX
plt.contourf(XX,YY,ZZ)

plt.plot(x,np.ones(len(x))*0.0)

plt.show()
```

```python
f=interpolate.RectBivariateSpline(x, y, ZZ.T)
xslice = np.linspace(-10*np.pi,10*np.pi,1000)
zslice = f(xslice,np.array([0.0]))

plt.plot(xslice,zslice)
#print xslice
#print zslice
plt.plot(xslice,np.cos(xslice),'--k')
plt.show()
```
